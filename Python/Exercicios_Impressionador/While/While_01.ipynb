#### 1. Faça um programa que peça uma nota, entre zero e dez. Mostre uma mensagem caso o valor seja inválido e continue pedindo até que o usuário informe um valor válido.

```python
while True:
    nota = int(input('Digite uma nota entre 0 e 10: '))
    
    if nota > 10 or nota < 0:
        print('Valor digitado é inválido')
    else:
        print('Nota correta')
        break
```

#### 2. Faça um programa que leia um nome de usuário e a sua senha e não aceite a senha igual ao nome do usuário, mostrando uma mensagem de erro e voltando a pedir as informações.

```python
while True:
    nome = input('Nome: ')
    senha = input('Senha: ')

    if senha in nome:
        print('A senha não pode ser igual ao nome')
    else:
        print('Cadastro realizado com sucesso')
        break
```

#### 3. Faça um programa que leia e valide as seguintes informações (e para cada uma delas, continue pedindo a informação até o usuário inserir corretamente):
##### Nome: maior que 3 caracteres;
##### Idade: entre 0 e 150;
##### Salário: maior que zero;
##### Sexo: 'f' ou 'm';
##### Estado Civil: 's', 'c', 'v', 'd';

```python
while True:
    nome = input('Nome: ')

    if len(nome) < 4:
        print('Seu nome precisa ter mais que 3 caracteres')
        continue
    break

while True:
    idade = int(input('Idade: '))

    if idade < 0 or idade > 150: # idade < 0 and idade > 150
        print('Sua idade deve estar entre 0 e 150 anos')
        continue
    break

while True:
    salario = int(input('Salário: '))

    if salario < 0:
        print('Salário precisa ser maior que zero')
        continue
    break

while True:
    sexo = input('Sexo: [M]asculino ou [F]eminino: ').upper()

    if 'F' not in sexo and 'M' not in sexo:
        print('Digite "M" para masculino ou "F" para feminino')
        continue
    
    break

while True:
    estado_civil = input(\
        'Digite seu estado civil:\n[S]olteiro\n[C]asado\n[V]iuvo\n[D]disquitado'\
            ).upper()
    
    if 'S' not in estado_civil and 'C' not in estado_civil \
        and 'V' not in estado_civil and 'D' not in estado_civil:
        print("Valor inválido")
        continue
    break

print("Cadastro: ")
print(f'Nome: {nome}\nIdade: {idade}\nSalário R$ {salario:.2f}')
print(f'Sexo: {sexo}\nEstado civil: {estado_civil}')
```

#### 4. Supondo que a população de um país A seja da ordem de 80000 habitantes com uma taxa anual de crescimento de 3% e que a população de B seja 200000 habitantes com uma taxa de crescimento de 1.5%. Faça um programa que calcule e escreva o número de anos necessários para que a população do país A ultrapasse ou iguale a população do país B, mantidas as taxas de crescimento.

```python
pais_A = 80000
pais_B = 200000
contador = 0

while True:
    pais_A += pais_A * 0.03
    pais_B += pais_B * 0.015
    
    contador += 1

    if pais_B <= pais_A:
        print('País "A" Ultrapassou ou igualou a população com o país "B"')
        print(f'Habitantes no País "A": {pais_A:.2f}')
        print(f'Habitantes no País "B": {pais_B:.2f}')
        print(f'Foram necessários {contador} anos')
        break

```

#### 6. Faça um programa que peça para o usuário inserir o faturamento dos últimos 5 meses (individualmente) e informe o maior faturamento

```python
contador = 1
faturamento_maximo = 0

while contador < 6:
    faturamento = float(input('Insira o faturamento do {contador}° mês: '))

    contador += 1

    if faturamento_maximo < faturamento:
        faturamento_maximo = faturamento

print(f'Maior faturamento: {faturamento_maximo:.2f}')
```

#### 7. Faça um programa que peça para o usuário inserir o faturamento dos últimos 5 meses (individualmente) e informe o faturamento total (soma) e o faturamento médio por mês (média).

```python
contador = 1
faturamento_soma = 0

while contador < 6:
    faturamento = float(input('Faturamento do {}° mês: '.format(contador)))
    faturamento_soma += faturamento
    contador += 1

faturamento_medio = faturamento_soma / 5

print(f'Faturmamento total: R$ {faturamento_soma:.2f}')
print(f'Faturamento médio: R$ {faturamento_medio:.2f}')

```

#### 8. Faça um programa que consiga categorizar a idade das equipes de uma empresa. Faça um programa que peça para n pessoas a sua idade, ao final o programa devera verificar se a média de idade da equipe varia entre 0 e 25 (jovem) ,26 e 60 (sênior) e maior que 60 (idosa); e então, dizer se a equipe é jovem, sênior ou idosa, conforme a média calculada.

```python
num = int(input('Número de funcionários: '))

print('Digite suas idade: ')

contador = num
soma_idade = 0

while contador > 0:
    idade = int(input('Idade: '))
    contador -= 1
    soma_idade += idade


media = soma_idade / num

if media >= 0 and media <= 25:
    print(f'Idade Média: {media}\nEquipe Joven')
elif media > 25 and media < 60:
    print(f'Idade Média: {media}\nEquipe Sênior')
else:
    print(f'Idade Média: {media}\nEquipe Idosa')
```

#### 9. Numa eleição existem três candidatos. Faça um programa que peça o número total de eleitores. Peça para cada eleitor votar e ao final mostrar o número de votos de cada candidato.

```python
from random import randint
 
eleitores = 156454011 # Número de votantes no brasil sengundo TSE

bolsonaro = 0
lula = 0
ciro = 0

contador = eleitores

print('Candidatos\n21 - Bolsonaro\n12 - Ciro Gomes\n13 - Lula')

while contador > 0:
    
    voto = randint(0, 2)

    match voto:
        case 0:
            bolsonaro += 1
            contador -= 1
            continue
        case 1:
            ciro += 1
            contador -= 1
            continue
        case 2:
            lula += 1
            contador -= 1
            continue
        case _:
            print('Candidato inexistente')
            

print('\n')



if lula < bolsonaro > ciro:
    print(f'Bolsonaro ganhou com {bolsonaro} votos')
    print(f'Ciro gomes: {ciro} votos')
    print(f'Lula: {lula} votos')
elif bolsonaro < ciro > lula:
    print(f'Ciro Gomes ganhou com {ciro} votos')
    print(f'Bolsonaro: {bolsonaro} votos')
    print(f'Lula: {lula} votos')
elif bolsonaro < lula > ciro:
    print(f'Lula ganhou com {lula} votos')
    print(f'Ciro gomes: {ciro} votos')
    print(f'Bolsonaro: {bolsonaro} votos')
elif bolsonaro == lula or lula == ciro or ciro == bolsonaro:
    print('Eleições empatadas ')
```

#### 11. O Sr. Manoel Joaquim possui uma grande loja de artigos de R\\$ 1,99, com cerca de 10 caixas. Para agilizar o cálculo de quanto cada cliente deve pagar ele desenvolveu um tabela que contém o número de itens que o cliente comprou e ao lado o valor da conta. Desta forma a atendente do caixa precisa apenas contar quantos itens o cliente está levando e olhar na tabela de preços. Você foi contratado para desenvolver o programa que monta esta tabela de preços, que conterá os preços de 1 até 50 produtos, conforme o exemplo abaixo:
<pre>
Lojas Quase Dois - Tabela de preços
1 - R$ 1.99
2 - R$ 3.98
...
50 - R$ 99.50
</pre>

```python
quantidade = int(input('Número de produtos: '))
preço = 1.99
valor_pagar = 0

while quantidade > 0:
    valor_pagar += preço
    quantidade -= 1

print(f'Valor a cborar: R$ {valor_pagar:.2f}')

```

#### 12. Um funcionário de uma empresa recebe aumento salarial anualmente: Sabe-se que:
<pre>
Esse funcionário foi contratado em 1995, com salário inicial de R$ 1.000,00;
Em 1996 recebeu aumento de 1,5% sobre seu salário inicial;
A partir de 1997 (inclusive), os aumentos salariais sempre correspondem ao dobro do percentual do ano anterior. Faça um programa que determine o salário desse funcionário em 2003. 
</pre>

```python
salario = 1000
ano = 1995
aumento = 1.5

while ano < 2000:
    ano += 1
    salario = salario * aumento / 100 + salario
    aumento *= 2

print(f'\nSalário em 2000: {salario:.2f}')
```

#### 13. O cardápio de uma lanchonete é o seguinte:
<pre>
Especificação   Código  Preço
Cachorro Quente 100     R$ 1,20
Bauru Simples   101     R$ 1,30
Bauru com ovo   102     R$ 1,50
Hambúrguer      103     R$ 1,20
Cheeseburguer   104     R$ 1,30
Refrigerante    105     R$ 1,00
Faça um programa que leia o código dos itens pedidos e as quantidades desejadas. Calcule e mostre o valor a ser pago por item (preço * quantidade) e o total geral do pedido. Considere que o cliente deve informar quando o pedido deve ser encerrado.
</pre>

```python
cardapio = [
    ('cachorro_quente', 100, 1.20),
    ('bauru_simples', 101, 1.30),
    ('bauru_ovo', 102, 1.50),
    ('haburguer', 103, 1.20),
    ('cheeseburguer', 104, 1.30),
    ('refrigerente', 105, 1.00),
]

preco = 0
lista_produtos = []

while True:
    codigo = int(input('código do produto: '))

    if codigo == 0:
        print('Pedido finalizado')
        break

    for produto in cardapio:
        if codigo in produto:
            nome, codigo, valor = produto
            lista_produtos.append(nome)
            preco += valor
    

print('\nPedido: ')
for name in lista_produtos:
    print(name)
print(f'\nValor a pagar: {preco:.2f}')
```
